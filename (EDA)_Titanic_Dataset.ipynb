```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("titanic")
```

```python
print("First 5 rows of dataset:\n", df.head(), "\n")
print("Dataset Info:\n")
df.info()
print("\nSummary Statistics:\n", df.describe(include='all'))
```

```python
print("\nMissing values in each column:\n", df.isnull().sum())

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap="coolwarm")
plt.title("Missing Values in Titanic Dataset")
plt.show()
```

```python
df.hist(bins=20, figsize=(12,8), color='skyblue')
plt.suptitle("Distribution of Numerical Features", fontsize=15)
plt.show()

categorical_features = ['sex', 'class', 'embarked', 'alive']
for col in categorical_features:
    plt.figure(figsize=(5,3))
    sns.countplot(data=df, x=col, palette="viridis")
    plt.title(f"Count of {col}")
    plt.show()
```

```python
# Survival rate by gender
plt.figure(figsize=(5,3))
sns.barplot(x="sex", y="survived", data=df, palette="Set2")
plt.title("Survival Rate by Gender")
plt.show()

# Survival rate by class
plt.figure(figsize=(5,3))
sns.barplot(x="class", y="survived", data=df, palette="Set1")
plt.title("Survival Rate by Passenger Class")
plt.show()

# Age distribution by survival
plt.figure(figsize=(6,4))
sns.boxplot(x="survived", y="age", data=df, palette="pastel")
plt.title("Age Distribution vs Survival")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
